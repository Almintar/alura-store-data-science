### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Verificar integridad de la informacion
print(tienda4.info())
```

```python
Tiendas = {'tienda_1':tienda, 'tienda_2':tienda2, 'tienda_3':tienda3, 'tienda_4':tienda4}
```

#1. Análisis de facturación



```python
for key, value in Tiendas.items():
    print(f'Las ventas totales para la {key} son: {value["Precio"].sum()}')
```

# 2. Ventas por categoría

```python
for key, value in Tiendas.items():
    print(f'Las 3 categorías mas populares para la {key} son: {value["Categoría del Producto"].value_counts().head(3)}')
```

# 3. Calificación promedio de la tienda


```python
for key, value in Tiendas.items():
    print(f'La calificación promedio la {key} son: {round(value["Calificación"].mean(), 2)}')
```

# 4. Productos más y menos vendidos

```python
for key, value in Tiendas.items():
  productos_contados = value["Producto"].value_counts()
  print(f'Los 3 productos mas vendidos para la {key} son: {productos_contados.head(3)}')
  print(f'Los 3 productos menos vendidos para la {key} son: {productos_contados.tail(3)}')
  print('\n')
```

# 5. Envío promedio por tienda

```python
for key, value in Tiendas.items():
    print(f'El costo promedio de envío para la {key} son: {round(value["Costo de envío"].mean(), 2)}')
```

# 6. Resultados

```python
resumen_tiendas = pd.DataFrame()
for key, value in Tiendas.items():
    # ventas totales
    vt = value['Precio'].sum()
    # top 3 de categorias más vendidos
    vxc = value['Categoría del Producto'].value_counts().head(3).index.values
    # Calificación por tienda
    cxt = round(value['Calificación'].mean(), 2)
    # Productos mas vendidos
    pmasv = value['Producto'].value_counts().head(3).index.values
    # Productos menos vendidos
    pmenosv = value['Producto'].value_counts().tail(3).index.values
    # Costo de envío por tienda
    cenv = round(value['Costo de envío'].mean(), 2)
    # costos totales de envio
    vtenv = value['Costo de envío'].sum()

    tienda = pd.DataFrame({
        'Tienda': key,
        'venta totales': [vt],
        'top 3 categoria': [vxc],
        'calificación promedio': [cxt],
        'top productos mas vendidos': [pmasv],
        'top productos menos vendidos': [pmenosv],
        'envio promnedio': [cenv],
        'Costo de envío': [vtenv],
        'Cantidad de vendedores': [len(value['Vendedor'].unique())]
        })

    resumen_tiendas = pd.concat([resumen_tiendas, tienda], axis = 0)

resumen_tiendas['Ganancias'] = resumen_tiendas['venta totales'] - resumen_tiendas['Costo de envío']

resumen_tiendas
```

```python
# 1. Barras: Ventas totales por tienda
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.bar(Tiendas.keys(), resumen_tiendas["venta totales"], color='skyblue')
plt.title("Ventas Totales por Tienda")
plt.ylabel("COP")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
# 2. Columna-Linea: Ventas totales vs Costo total de envio

resumen_tiendas = resumen_tiendas.sort_values(['Tienda'])
categorias = resumen_tiendas['Tienda']
x = np.arange(len(categorias))  # posiciones en el eje x

# Crear figura y eje
fig, ax1 = plt.subplots()

# Barras (eje primario)
ax1.bar(x, resumen_tiendas['venta totales'], color='skyblue', label='Barras')
ax1.set_ylabel('Ventas Totales')
ax1.set_xticks(x)
ax1.set_xticklabels(categorias)

# Línea (eje secundario)
ax2 = ax1.twinx()
ax2.plot(x, resumen_tiendas['Costo de envío'], color='red', marker='o', label='Línea')
ax2.set_ylabel('Costo total de envíos')

# Títulos y leyendas
fig.suptitle('Ventas vs Envios por tienda')
ax1.set_xlabel('Tiendas')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Crear gráfico de barras horizontales
plt.figure(figsize=(8, 5))
plt.barh(resumen_tiendas['Tienda'], resumen_tiendas['calificación promedio'], color='skyblue')
plt.xscale('log')  # Solo si tiene sentido logarítmico

# Etiquetas y título
plt.xlabel('Valor de Barras')
plt.ylabel('Tiendas')
plt.title('Calificacion por Tienda')

plt.tight_layout()
plt.show()
```

```python
# Crear gráfico de barras horizontales
plt.figure(figsize=(8, 5))
plt.barh(resumen_tiendas['Tienda'], resumen_tiendas['Ganancias'], color='skyblue')
#plt.xscale('log')  # Solo si tiene sentido logarítmico

# Etiquetas y título
plt.xlabel('Valor de Barras')
plt.ylabel('Tiendas')
plt.title('Ganancias netas')

plt.tight_layout()
plt.show()
```
